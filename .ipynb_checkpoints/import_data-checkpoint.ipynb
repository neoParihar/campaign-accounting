```python
import pandas as pd
import numpy as np
import os
from dotenv import load_dotenv, dotenv_values
import mysql.connector
```

```python
load_dotenv()
try:
    conn = mysql.connector.connect(
        user = os.getenv("DB_USER"),
        host = os.getenv("DB_HOST"),
        password = os.getenv("DB_PASSWORD"),
        database = os.getenv("DATABASE")
    )
    if conn.is_connected():
        print("Connected to MySQL Database")
except Error as e:
    print("Error while connecting:", e)
```

```python
candidates = pd.read_csv("data/candidate_master.txt", sep="|", header=None, 
                         names=["cand_id", "cand_name", "cand_pty_affiliation", "cand_election_yr", 
                                "cand_office_st", "cand_office", "cand_office_district", "cand_ici", 
                                "cand_status", "cand_pcc", "cand_st1", "cand_st2", "cand_city", 
                                "cand_st", "cand_zip"])
candidate_db = candidates[["cand_id", "cand_name", "cand_pty_affiliation", "cand_election_yr", 
                           "cand_office_st", "cand_office", "cand_office_district", "cand_ici", 
                           "cand_status", "cand_pcc"]]
cand_data = [tuple(None if pd.isna(x) else x for x in row) for row in candidate_db.to_numpy()]
```

```python
committees = pd.read_csv("data/committee_master26.txt", sep="|", header=None, 
                         names=["comm_id", "comm_name", "tres_name", "comm_st1",
                               "comm_st2", "comm_city", "comm_state", "comm_zip",
                               "comm_design", "comm_type", "comm_pty", "comm_filing",
                               "org_interest", "org_name", "cand_id"])
committees_db = committees[["comm_id", "comm_name", "comm_design", "comm_type",
                           "comm_pty", "comm_filing", "org_interest", "org_name",
                           "cand_id"]]
committees_data =[tuple(None if pd.isna(x) else x for x in row) for row in committees_db.to_numpy()]
```

```python
candcomm = pd.read_csv("data/candcomm26.txt", sep="|", header=None,
                       names=["cand_id", "cand_elect_yr", "FEC_elect_yr", "comm_id", 
                              "comm_type", "comm_design", "link_id"])
candcomm_db = candcomm[["link_id", "cand_id", "comm_id"]]
candcomm_data = [tuple(None if pd.isna(x) else x for x in row) for row in candcomm_db.to_numpy()]
```

```python
cursor = conn.cursor()
#cursor.execute("ALTER TABLE Committees DROP CONSTRAINT Committees_ibfk_1;")
cursor.execute("DROP TABLE IF EXISTS CandidateCommittee")
cursor.execute("DROP TABLE IF EXISTS Candidates")
cursor.execute("DROP TABLE IF EXISTS Committees")
```

```python
cursor.execute("""
               CREATE TABLE Candidates (
                   CandID VARCHAR(9) PRIMARY KEY,
                   CandName VARCHAR(200),
                   CandParty VARCHAR(3),
                   CandElectYr INT,
                   CandStateAbbr VARCHAR(2),
                   CandOffice VARCHAR(1),
                   CandOfficeDist INT,
                   CandIncumbent VARCHAR(1),
                   CandStatus VARCHAR(1),
                   CandPCC VARCHAR(9)
                );
                """)
cursor.execute("""
               CREATE TABLE Committees (
                   CommID VARCHAR(9) PRIMARY KEY,
                   CommName VARCHAR(200),
                   Designation VARCHAR(1),
                   CommType VARCHAR(1),
                   CommParty VARCHAR(3),
                   FilingFreq VARCHAR(1),
                   InterestGrpCategory VARCHAR(1),
                   ConnOrgName VARCHAR(200),
                   CandID VARCHAR(9),
                   FOREIGN KEY (CandID) REFERENCES Candidates(CandID)
               );
                """)
cursor.execute("ALTER TABLE Candidates ADD CONSTRAINT fk_pcc FOREIGN KEY(CandPCC) REFERENCES Committees(CommID);")
cursor.execute("""
                CREATE TABLE CandidateCommittee (
                    LinkageID INT PRIMARY KEY,
                    CandID VARCHAR(9),
                    CommID VARCHAR(9),
                    FOREIGN KEY (CandID) REFERENCES Candidates(CandID),
                    FOREIGN KEY (CommID) REFERENCES Committees(CommID)
                );
                """)
conn.commit()
```

```python
cand_query = """
        INSERT INTO Candidates(CandID, CandName, CandParty, CandElectYr, CandStateAbbr, 
                                CandOffice, CandOfficeDist, CandIncumbent, CandStatus, CandPCC)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s);
        """
cursor.executemany(cand_query, cand_data)
conn.commit()
Programmi
```
